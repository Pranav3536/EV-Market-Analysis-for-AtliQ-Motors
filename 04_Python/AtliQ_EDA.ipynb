# Section 3 — Missing Values Analysis

```python
# Check missing values in each column
print("Missing Values in Each Column:")
print(df.isnull().sum())


# Calculate missing values in percentage
missing_percentage = (df.isnull().sum() / len(df)) * 100
print("\nMissing Values Percentage:")
print(missing_percentage.round(2))
```

## Observation

- We checked the dataset for missing values in each column.
- The number and percentage of missing values help us identify which columns need attention.

# Section 4 — Duplicate Records Analysis

```python
# Check duplicate records
duplicate_rows = df.duplicated().sum()
print("Total Duplicate Rows:", duplicate_rows)


# Display duplicate records (if any)
if duplicate_rows > 0:
    print("\nDuplicate Records:")
    display(df[df.duplicated()])
else:
    print("\nNo duplicate records found.")
```

## Observation

- We checked the dataset for duplicate records.
- Duplicate rows can lead to incorrect analysis and inaccurate results.
- If duplicate records are found, they will be removed during the Data Cleaning phase.

# Section 5 — Data Quality Check

```python
# Check data types of all columns
print("Data Types:")
print(df.dtypes)


# Check unique values in the Maker column
print("\nUnique Maker Names:")
print(df["Maker"].unique())


# Check for extra spaces in the Maker column
extra_spaces = df["Maker"] != df["Maker"].str.strip()
print("\nRows with Extra Spaces in Maker Column:")
print(extra_spaces.sum())


# Display rows having extra spaces
display(df[extra_spaces])
```

## Observation

- We reviewed the data types of all columns.
- The unique values in the **Maker** column were inspected to identify any inconsistent text formatting.
- Some manufacturer names contain leading or trailing spaces.
- These extra spaces can create duplicate categories during analysis.
- The **Maker** column will be cleaned by removing unnecessary spaces in the Data Cleaning section.

# Section 6 — Data Cleaning

```python
# Fill missing values with 0
df["Electric_vehicles_sold_bystates"] = df["Electric_vehicles_sold_bystates"].fillna(0)


# Remove extra spaces from the Maker column
df["Maker"] = df["Maker"].str.strip()


# Convert the column to integer
df["Electric_vehicles_sold_bystates"] = df["Electric_vehicles_sold_bystates"].astype(int)


# Reset the index
df.reset_index(drop=True, inplace=True)


# Check missing values after cleaning
print("Missing Values After Cleaning:")
print(df.isnull().sum())


# Display first 5 rows after cleaning
display(df.head())
```

## Observation

- Missing values in the **Electric_vehicles_sold_bystates** column have been replaced with **0**.
- Extra spaces were removed from the **Maker** column to ensure consistent manufacturer names.
- The **Electric_vehicles_sold_bystates** column was converted to an integer data type after handling missing values.
- The dataset index was reset to keep the data organized.
- No duplicate records were removed because no duplicate rows were found during the data inspection.
- The dataset is now clean and ready for feature engineering.

# Section 7 — Feature Engineering

```python
# Convert Date column to datetime format
df["Date"] = pd.to_datetime(df["Date"])


# Create Year column
df["Year"] = df["Date"].dt.year


# Create Month column
df["Month"] = df["Date"].dt.month_name()


# Create Month Number column
df["Month_Number"] = df["Date"].dt.month


# Create EV Penetration (%) column
df["EV_Penetration_%"] = (
    df["Electric_vehicles_sold_bystates"] /
    df["Total_vehicles_sold"]
) * 100
df["EV_Penetration_%"] = df["EV_Penetration_%"].round(2)


# Create EV Penetration Category column
df["EV_Penetration_Category"] = "Low"
df.loc[df["EV_Penetration_%"] >= 5, "EV_Penetration_Category"] = "Medium"
df.loc[df["EV_Penetration_%"] >= 10, "EV_Penetration_Category"] = "High"


# Display first 5 rows
display(df.head())
```

## Observation

- The **Date** column was converted into datetime format for easier date-based analysis.
- Three new columns (**Year**, **Month**, and **Month_Number**) were created from the Date column.
- The **EV_Penetration_%** column was calculated using the ratio of electric vehicles sold to total vehicles sold and rounded to two decimal places.
- A new **EV_Penetration_Category** column was created to classify EV adoption into **Low**, **Medium**, and **High** categories.
- The dataset now contains additional business-friendly features that will be useful for Power BI dashboards and further analysis.

# Section 8 — Final Dataset & Export

```python
# Arrange the columns
df = df[
    [
        "Date",
        "Fiscal_year",
        "Quarter",
        "Year",
        "Month",
        "Month_Number",
        "State",
        "Maker",
        "Vehicle_category",
        "Electric_vehicles_sold_maker",
        "Electric_vehicles_sold_bystates",
        "Total_vehicles_sold",
        "EV_Penetration_%",
        "EV_Penetration_Category"
    ]
]


# Display the final dataset
print("Final Dataset:")
display(df.head())


# Export the cleaned dataset
output_path = r"C:\Users\Pranav\EV_Project\Datasets\Atliq_ev_market_clean.xlsx"

df.to_excel(output_path, index=False)

print("Dataset exported successfully.")
print("Location:", output_path)
```
